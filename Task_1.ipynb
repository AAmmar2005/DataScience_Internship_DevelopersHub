```python
# Importing tools for data analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# For graphs nd charts 
sns.set_theme(style="whitegrid")
vibrant_palette = "magma" 

```

## 1. Data Loading and Initial Inspection
check dataset into a DataFrame .
using info(), head().

```python

# Loading dataset 
iris_df = sns.load_dataset('iris')

# Checking rows of the columns
print("Preview of the Dataset:")
print(iris_df.head())

#  shape and column
print(f"\nTotal Records: {iris_df.shape[0]} | Total Features: {iris_df.shape[1]}")
print("\nDataset Information:")
iris_df.info()
```


## 2. Data Cleaning & Preparation



```python
# Checking null values
print("Checking for missing data:")
print(iris_df.isnull().sum())

# Checking duplicate records
duplicates = iris_df.duplicated().sum()
print(f"\nNumber of duplicate rows found: {duplicates}")

# Removing duplicates from the dataset
if duplicates > 0:
    iris_df = iris_df.drop_duplicates()
    print("Action: Duplicates removed successfully.")


print("\nQuick Statistical Summary:")
print(iris_df.describe())
```

## 3. Exploratory Data Analysis (Visualizing Patterns)


### **A. Scatter Plot**



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species', palette=vibrant_palette, s=70)
plt.title('Sepal Length vs. Sepal Width (Species Breakdown)', loc='center', fontsize=15)
plt.show()
```

### **B. Histogram: Petal Length**



```python
plt.figure(figsize=(8, 5))
sns.histplot(iris_df['petal_length'], bins=20, kde=True, color="#360b31")
plt.title('Distribution Analysis of Petal Length', fontsize=15)
plt.xlabel('Petal Length (cm)')
plt.show()
```

### **C. Box Plot: Outlier Detection and Data Spread**
I use a box plot , for check numerical feature.
 and check for unusual values.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_df, palette="viridis", orient="h")
plt.title('Feature Range and Outlier Check', fontsize=15)
plt.show()
```
